## A União Europeia

A União Europeia (UE) é um bloco econômico e político formado por países europeus. Atualmente fazem parte da União Europeia 28 países e a sede do bloco fica em Bruxelas, capital da Bélgica.

Os principais objetivos da UE são:

ajudar os países-membros no seu desenvolvimento econômico,  
dar mais igualdade política e econômica à Europa,  
melhorar as condições econômicas e de trabalho dos cidadãos europeus,  
reduzir as desigualdades econômicas e sociais entre todas as regiões que compõem o bloco,  
garantir a permanência de um ambiente pacífico e harmonioso em toda a Europa.  

## Explicação da hipótese e mecanismo

Um dos principais objetivos da União Europeia é ajudar os países-membros em seu desenvolvimento economico, logo, é razoável esperar que países participantes se recuperem com mais facilidade de uma potencial crise financeira.  


Este projeto tem como objetivo demonstrar, através de uma análise exploratória de dados, o efeito da participação na União Europeia no crescimento economico de um país em um cenário pós-crise-financeira.  
Para isso, será feita uma análise de variáveis como a renda per capita e a taxa de desemprego ao longo dos anos de 2008 a 2012 de países membros da união, comparando-as com países da europa e do mundo que não fazem parte da União Europeia.

```python
import pandas as pd
import matplotlib.pyplot as plt
import os
from numpy import arange
from numpy import percentile
import numpy as np
```

```python
gdp = pd.read_excel('gdp.xlsx')
unemp = pd.read_excel('unemployment.xlsx')
```

```python
gdp_2008 = gdp.loc[:, ["country", 2008]]
gdp_2009 = gdp.loc[:, ["country", 2009]]
gdp_2010 = gdp.loc[:, ["country", 2010]]
gdp_2011 = gdp.loc[:, ["country", 2011]]
gdp_2012 = gdp.loc[:, ["country", 2012]]
unemp_2008 = unemp.loc[:, ["country", 2008]]
unemp_2009 = unemp.loc[:, ["country", 2009]]
unemp_2010 = unemp.loc[:, ["country", 2010]]
unemp_2011 = unemp.loc[:, ["country", 2011]]
unemp_2012 = unemp.loc[:, ["country", 2012]]
```

```python
gdp_2008_idx = gdp_2008.set_index("country")
gdp_2009_idx = gdp_2009.set_index("country")
gdp_2010_idx = gdp_2010.set_index("country")
gdp_2011_idx = gdp_2011.set_index("country")
gdp_2012_idx = gdp_2012.set_index("country")
unemp_2008_idx = unemp_2008.set_index("country")
unemp_2009_idx = unemp_2009.set_index("country")
unemp_2010_idx = unemp_2010.set_index("country")
unemp_2011_idx = unemp_2011.set_index("country")
unemp_2012_idx = unemp_2012.set_index("country")
```

```python
df_2008 = gdp_2008_idx.join(unemp_2008_idx, how = 'inner', lsuffix = ': GDP', rsuffix = ': Desemprego')
df_2009 = gdp_2009_idx.join(unemp_2009_idx, how = 'inner', lsuffix = ': GDP', rsuffix = ': Desemprego')
df_2010 = gdp_2010_idx.join(unemp_2010_idx, how = 'inner', lsuffix = ': GDP', rsuffix = ': Desemprego')
df_2011 = gdp_2011_idx.join(unemp_2011_idx, how = 'inner', lsuffix = ': GDP', rsuffix = ': Desemprego')
df_2012 = gdp_2012_idx.join(unemp_2012_idx, how = 'inner', lsuffix = ': GDP', rsuffix = ': Desemprego')

frames = [df_2009,df_2010,df_2011,df_2012]

data = df_2008.join(frames,how='inner')

data


```

```python
UE = ('Germany','Austria','Belgium','Bulgaria','Cyprus','Croatia','Denmark','Slovakia','Slovenia','Spain','Estonia','Finland','France','Greece','Netherlands','Hungary','Ireland','Italy','Latvia','Lithuania','Luxemburg','Malta','Poland','Portugal','Czech Republic','Romenia','Sweden')
```

```python
data["União Europeia"] = "Não Pertence"
for i in UE:
    data.loc["{}".format(i), "União Europeia"] = "Pertence"
data
```

```python
#Remove entries com valor NaN
data = data.dropna(how='any')

```

```python
data_UE = data.loc[data["União Europeia"] == "Pertence"]
data_UE
```

## Comparação de nações similares

## - União Europeia

```python
values_gdp = ['2008: GDP','2009: GDP','2010: GDP','2011: GDP','2012: GDP']
values_desemprego = ['2008: Desemprego','2009: Desemprego','2010: Desemprego','2011: Desemprego','2012: Desemprego']

gdp_UE = data_UE.loc[:,values_gdp]
desemprego_UE = data_UE.loc[:,values_desemprego]
gdp_UE
```

```python
gdpUE2008 = data_UE.loc[:,"2008: GDP"]
gdpUE2009 = data_UE.loc[:,"2009: GDP"]
gdpUE2010 = data_UE.loc[:,"2010: GDP"]
gdpUE2011 = data_UE.loc[:,"2011: GDP"]
gdpUE2012 = data_UE.loc[:,"2012: GDP"]

print('Renda Média da União Europeia entre os anos de 2008 e 2012:')
print('Renda Média no ano de 2008:', gdpUE2008.mean())
print('Renda Média no ano de 2009:', gdpUE2009.mean())
print('Renda Média no ano de 2010:', gdpUE2010.mean())
print('Renda Média no ano de 2011:', gdpUE2011.mean())
print('Renda Média no ano de 2012:', gdpUE2012.mean())
```

```python
desemprego_UE
```

```python
gdp_UE.plot.bar(x=None,y=values_gdp, figsize=(20, 8), color=('b','g','r','orange','gold'))
desemprego_UE.plot.bar(x=None,y=values_desemprego, figsize=(20,8), color=('b','g','r','orange','gold'))
```

    Analisando os gráficos a cima, é possível notar uma heterogeneidade no crescimento econômico de países membros da União Europeia. Enquanto países como Alemanha, Dinamarca e Polônia apresentam um crescimento significativo ao longo dos anos de 2008 e 2012, Irlanda, Holanda e Eslovenia apresentam dificuldade para se recuperar da crise de 2008, o que indica que apenas fazer parte da União Europeia não é o suficiente para um país se recuperar economicamente após uma crise financeira mundial.
